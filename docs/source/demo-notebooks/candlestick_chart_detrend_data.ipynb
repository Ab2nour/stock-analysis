# How detrended data behave with candlestick chart

The whole data are detrended using close price (chosen arbitrarily amongst open / close / high / low).  
This way, the detrended data can be plotted on candlestick chart.    
Otherwise, open / close / high / low prices may cross each other.  

```python
import mplfinance as mpl
import pandas as pd

from src.utils import init_notebook
```

```python
init_notebook()
```

Choose detrend method

```python
detrend_model_name = "LinearMADetrend"
model_options = "window-100"
```

```python
raw_data_folder = "data/raw_data"
processed_data_folder = (
    f"data/processed_data/detrend_data/{detrend_model_name}/{model_options}"
)
stock_name = "AAPL"
```

```python
df = pd.read_csv(
    f"{raw_data_folder}/{stock_name}.csv", parse_dates=["Date"], index_col="Date"
)
print(f"{df.shape = }")

df_detrend = pd.read_csv(
    f"{processed_data_folder}/{stock_name}.csv",
    parse_dates=["Date"],
    index_col="Date",
)
```

```python
df_6_months = df[df.index > pd.Timestamp("2021-06-01")]
df_detrend_6_months = df_detrend[df.index > pd.Timestamp("2021-06-01")]
```

```python
mpl.plot(
    df_6_months,
    type="candle",
    title=f"Cours de l'action {stock_name}",
    style="yahoo",
)
mpl.plot(
    df_detrend_6_months,
    type="candle",
    title=f"Cours de l'action {stock_name} sans la tendance",
    style="yahoo",
)
```
