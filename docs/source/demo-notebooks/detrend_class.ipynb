# How to use detrend classes

```python
import pandas as pd

from src.functions.detrend import (
    BSplinesDetrend,
    ExponentialMADetrend,
    LinearMADetrend,
    LinearRegressionDetrend,
    PolynomialRegressionDetrend,
)
from src.utils.misc import init_notebook
```

```python
init_notebook()
```

```python
# Lecture des donn√©es
data_folder = "data/raw"
stock_name = "AAPL"
df = pd.read_csv(
    f"{data_folder}/{stock_name}.csv", parse_dates=["Date"], index_col="Date"
)
y = df["Close"]
```

```python
models = [
    LinearRegressionDetrend(),
    PolynomialRegressionDetrend(order=5, n_segments=15),
    LinearMADetrend(window=500),
    ExponentialMADetrend(alpha=0.9),
    BSplinesDetrend(interval_length=50, degree=3),
]
```

```python
for detrend_model in models:
    y_fitted = detrend_model.fit(y)
    y_detrend = detrend_model.predict(y)
    detrend_model.fancy_plot(xticklabels=df.index.strftime("%Y-%m-%d"))
```
