```python
import yfinance as yf
import mplfinance as mpl
```

```python
data_folder = "../../data/raw_data"
```

```python
# change these to the stock you want to download
stock_name = "AAPL"
start = "2019-01-01"
end = "2021-12-31"
```

```python
df = yf.download(stock_name, start, end)
print(f"{df.shape = }")
```

```python
# plot the data to verify your needs
mpl.plot(df, type="candle", style="yahoo")
```

```python
df.to_csv(f"{data_folder}/{stock_name}.csv")
```
