# Imports et lecture des données

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import mplfinance as mpl
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
sns.set_theme()
```

## Lecture des données

```python
data_folder = "../data"
```

```python
stock_name = "AAPL"
detrend_stock_name = "AAPL._detrend_linearMA_window-100"
```

```python
df = pd.read_csv(
    f"{data_folder}/{stock_name}.csv", parse_dates=["Date"], index_col="Date"
)
print(f"{df.shape = }")

df_detrend = pd.read_csv(
    f"{data_folder}/detrend_data/{detrend_stock_name}.csv",
    parse_dates=["Date"],
    index_col="Date",
)
print(f"{df_detrend.shape = }")
```

# Analyse exploratoire

## Affichage des données

### Line plot

```python
df["Close"].plot(title=f"Cours de l'action {stock_name}", ylabel="Price")
df_detrend["Close"].plot()
```

### Candlestick chart (graphique en bougie

```python
df_3_mois = df[df.index < pd.Timestamp("2019-03-01")]
df_detrend_3_mois = df_detrend[df.index < pd.Timestamp("2019-03-01")]
```

```python
mpl.plot(
    df_3_mois,
    type="candle",
    title=f"Cours de l'action {stock_name}",
    style="yahoo",
)
mpl.plot(
    df_detrend_3_mois,
    type="candle",
    title=f"Cours de l'action {stock_name}",
    style="yahoo",
)
```

```python
mpl.plot(
    df,
    type="candle",
    title=f"Cours de l'action {stock_name}",
    style="yahoo",
    warn_too_much_data=10_000,
)
mpl.plot(
    df_detrend,
    type="candle",
    title=f"Cours de l'action {stock_name}",
    style="yahoo",
    warn_too_much_data=10_000,
)
```

```python
mpl.plot(
    df_3_mois,
    type="candle",
    mav=(3, 6, 9),
    title=f"Cours de l'action {stock_name}",
    style="yahoo",
)
```
